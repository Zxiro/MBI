```python
import json #import python json module 
import csv #import python csv module 
import numpy as np
import pandas as pd
#改成pandas處理csv
CsvData=pd.DataFrame(pd.read_csv('./stock_data/json/stock_0050.csv'))#After Reading csv put it in the dataframe
csv_data=open('./stock_data/json/stock_0050.csv')
outputData=open('./stock_data/training_data/0050_train_data.csv','w',newline="")
ls=[]
print(CsvData)
for line in csv_data:
    line=line.replace("\n","") #將換行換為空
    #strip
    ls.append(line.split(",")) #每遇到","即分割
csv_data.close()
#print("It's csv:   ",ls,"\n")

for i in range(1,len(ls)):
    ls[i]=dict(zip(ls[0],ls[i]))
    #zip 將兩個對應的參數組合起來 
    #dict()創建為字典

json_data=json.dumps(ls[1:])
#print(json_data,"\n\n")
#data=json.loads(json_data)#datatype json to dict

SelectFeature=['date','open','high','low','volume','close']#All the var for the model
Feature=[]
train_x=[]
train_y=[]
result=[]
tmp=[]

#if keys are different
#for dic in json_data:
# for i in list(dic.keys()): #for python3  it's required that u have to use list(dct_keys) to get the values
# keys.append(i)'''
#print(keys,"\n\n")# all the keys in json

for dic in eval(json_data):
    for i in range(0,6):
        if i!=5:
            tmp.append(dic.get(SelectFeature[i]))
        else:
            train_y.append(dic.get(SelectFeature[i]))
for i in range(0,len(tmp)-1,5):
    train_x.append(tmp[i:i+5])
    
for i in range(0,len(train_x)):
    if float(train_x[i][1])-float(train_y[i])>0:
        result.append(1)#close<open
    elif float(train_x[i][1])-float(train_y[i])==0:
        result.append(0)#close==open
    elif float(train_x[i][1])-float(train_y[i])<0:
        result.append(2)#close>open


print("Train_x:  ",train_x)
print("\n\nTrain_y:  ",train_y) 
#print("\n\nResult: ",result)
    
writer=csv.writer(outputData)#寫入 csv
writer.writerow(train_x)
writer.writerow(train_y)
writer.writerow(result)
outputData.close()
```
