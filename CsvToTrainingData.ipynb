```python
import json 
import csv
import datetime#import datetime module convert data to datetime64(ns)
import numpy as np
import pandas as pd#import pandas module(Better than csv module!)

global train_x
train_x=[]
def GenerateTrainX(n,o,h,l,c,v):#n for numbers /d for days /o for open /h for high /l for low /c for close /v for volume
    fliter=np.array([bool(o),bool(h),bool(l),bool(c),bool(v)])
    Choice=Features[fliter]#type(Choice)=nparray
    #train_x=[]
    count=0
    TimeData=CsvData.set_index('date').resample('w')
    for i in TimeData:
        if count<n:
            #print(i[1])
            IsWeekOrNot=int(i[1].size)/6
            if IsWeekOrNot>=5:
                print(i[1].loc[ : ,Choice].values.tolist())
                train_x.append(i[1].loc[ : ,Choice].values.tolist())
                count+=1
            else:
                continue
        elif count>=n:
            break
    print("\n","Train_x:  ",train_x,"\n","Train_x.size:  ",len(train_x))

StockNum = input("stock input:")
CsvData=pd.DataFrame(pd.read_csv('./stock_data/json/stock_'+StockNum+'.csv'))#After Reading csv put it in the dataframe
outputData=open('./stock_data/training_data/'+StockNum+'TrainData.csv','w',newline="")
Features=np.array(['open','high','low','close','volume'])#All the var for the model
CsvData['date']=pd.to_datetime(CsvData['date'])#將date column 原本的datatype(object) 轉成 datetime64
CsvData["week"]=0
for i,j in zip(CsvData["date"],range(CsvData["date"].size)):
    CsvData["week"][j]= i.isocalendar()[2]

GenerateTrainX(150,1,0,1,0,0)

writer=csv.writer(outputData)#寫入 csv
writer.writerow(train_x)
outputData.close()
```
